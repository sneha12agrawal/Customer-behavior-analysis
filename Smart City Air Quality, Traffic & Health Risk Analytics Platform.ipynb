```python
import requests
import pandas as pd
import time

API_KEY =  "88904129fafefebcb6e90ccde11d98a31b2ce20bfd38ad5631098bce696750eb"
BASE_URL = "https://api.openaq.org/v3/locations"

headers = {
    "X-API-Key":  "88904129fafefebcb6e90ccde11d98a31b2ce20bfd38ad5631098bce696750eb"
}

def build_full_india_dataset():
    all_data = []
    page = 1
    limit = 100

    while True:
        print(f"Fetching page {page}...")

        try:
            response = requests.get(
                BASE_URL,
                headers=headers,
                params={
                    "country": "IN",
                    "limit": limit,
                    "page": page
                },
                timeout=15
            )

            if response.status_code != 200:
                print("Server error:", response.status_code)
                break

            results = response.json().get("results", [])

            if not results:
                print("No more data found.")
                break

            all_data.extend(results)
            print(f"Total records collected: {len(all_data)}")

            page += 1
            time.sleep(1.5)  

        except requests.exceptions.RequestException as e:
            print("Network error:", e)
            print("Retrying in 5 seconds...")
            time.sleep(5)

    df = pd.json_normalize(all_data)
    df.to_csv("full_india_air_quality_locations.csv", index=False)

    print("\nDATA COLLECTION COMPLETE")
    print("Saved as full_india_air_quality_locations.csv")
    print("Total rows:", len(df))

    return df

# RUN
df = build_full_india_dataset()
df.head()
```

```python
def clean_air_quality_data(df):
    # Drop missing values
    df = df.dropna(subset=["value"])

    # Remove invalid values
    df = df[df["value"] >= 0]

    # Filter health-related pollutants
    df = df[df["parameter"].isin(["pm25", "pm10", "no2", "o3", "co"])]

    # Convert datetime
    df["datetime"] = pd.to_datetime(df["datetime"], errors="coerce")

    # Remove duplicates
    df = df.drop_duplicates(subset=["location", "parameter", "datetime"])

    # Remove extreme outliers
    df = df[df["value"] < df["value"].quantile(0.99)]

    return df
```

```python
def clean_air_quality_data(df):
    # Drop useless columns
    drop_cols = ["distance", "datetimefirst", "datetimelast"]
    df = df.drop(columns=[c for c in drop_cols if c in df.columns])

    # Fill text columns
    df["name"] = df["name"].fillna("Unknown Station")
    df["locality"] = df["locality"].fillna("Unknown City")

    if "licenses" in df.columns:
        df["licenses"] = df["licenses"].fillna("Not Specified")

    # Fill and convert datetime
    if "datetimefirst_utc" in df.columns and "datetimelast_utc" in df.columns:
        df["datetimefirst_utc"] = df["datetimefirst_utc"].fillna(df["datetimelast_utc"])
        df["datetimefirst_utc"] = pd.to_datetime(df["datetimefirst_utc"], errors="coerce")
        df["datetimelast_utc"] = pd.to_datetime(df["datetimelast_utc"], errors="coerce")

    # Ensure valid coordinates
    df = df.dropna(subset=["coordinates_latitude", "coordinates_longitude"])

    return df

```
